```python
#importing packages
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
%matplotlib inline
```

```python
data = np.loadtxt("40.10_25_amplitude_cap.txt")
f = data[:,0]
Vi = data[:,1]
Vo = data[:,2]
g = 20*np.log10(Vo/Vi)
frange = np.arange(1e0,10e4,10)
```

```python
def phase(f,a,b,c):
    return np.rad2deg(a*np.arctan((b/(2*np.pi*f))+c))

popt, pcov = curve_fit(phase,f,g)
best = phase(frange,popt[0],popt[1],popt[2])
plt.semilogx(f,g,'bo',markersize=5, label = "Experimental")
plt.semilogx(frange, best, 'g-',linewidth=2, label = "Theoretical")
plt.axhline(linewidth=1, color='k')
plt.axvline(linewidth=1, color='k')
plt.title('Experimental Amplitude Response Plot [Capacitor]')
plt.ylabel('Gain [dB]')
plt.xlabel('Frequency [Hz]')
plt.legend(bbox_to_anchor=(0.035, 0.205), loc=2, borderaxespad=0.)
plt.axis([10**0,10**5,-10,30])
plt.grid(True)
plt.savefig('40.10_25_amplitude_cap_output.png')
```
