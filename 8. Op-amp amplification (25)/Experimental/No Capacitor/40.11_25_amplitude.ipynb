```python
#importing packages
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
%matplotlib inline
```

```python
data = np.loadtxt("40.11_25_amplitude.txt")
f = data[:,0]
Vi = data[:,1]
Vo = data[:,2]
g = 20*np.log10(Vo/Vi)
frange = np.arange(1,1e6,10)
```

```python
def gain(f,a,b,c,d):
    return a+b*np.log10(c+d*f)

popt, pcov = curve_fit(gain,f,g)
best = gain(frange,popt[0],popt[1],popt[2],popt[3])
plt.semilogx(f,g,'bo',markersize=5, label = "Experimental")
plt.semilogx(frange, best, 'g-',linewidth=2, label = "Theoretical")
plt.axhline(linewidth=1, color='k')
plt.axvline(linewidth=1, color='k')
plt.title('Experimental Amplitude Response Plot')
plt.ylabel('Gain [dB]')
plt.xlabel('Frequency [Hz]')
plt.legend(bbox_to_anchor=(0.035, 0.205), loc=2, borderaxespad=0.)
plt.axis([10**0,10**6,-10,30])
plt.grid(True)
plt.savefig('40.11_25_amplitude_output.png')
```
