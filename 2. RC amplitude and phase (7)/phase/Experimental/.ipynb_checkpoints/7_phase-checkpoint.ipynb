```python
#importing packages
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
%matplotlib inline
```

```python
data = np.loadtxt("7_phase.txt")
f = data[:,0]
p = data[:,1]
frange = np.arange(1,1e6,10)
R = 1e3
C = 1e-6
```

```python
def phase(f,a,b,c):
    return np.rad2deg(np.arctan(1/(2*np.pi*f*C*R)))

popt, pcov = curve_fit(phase,f,p)
best = phase(frange,popt[0],popt[1],popt[2])
plt.semilogx(f,p,'bo',markersize=5, label = "Experimental")
plt.semilogx(frange, best, 'g-',linewidth=1, label = "Theoretical")
plt.axhline(linewidth=1, color='k')
plt.axvline(linewidth=1, color='k')
plt.title('Experimental Phase Response Plot')
plt.ylabel('Phase [deg]')
plt.xlabel('Frequency [Hz]')
plt.axis([0,10**6,0,90])
plt.legend(bbox_to_anchor=(0.65, 0.95), loc=2, borderaxespad=0.)
plt.grid(True)
plt.savefig('7_phase.png')

print('At f = 1/2πRC the phase angle change of the transmitted signal is ~ {0:.6}'.format(phase(1/(2*np.pi*R*C),popt[0],popt[1],popt[2])),'degrees')
```
