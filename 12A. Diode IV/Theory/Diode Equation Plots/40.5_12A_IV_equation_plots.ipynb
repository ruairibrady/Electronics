```python
#importing packages
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
%matplotlib inline
```

```python
#variables
Is1 = 10e-12
Is2 = 10e-14
q = 1.602e-19
k = 1.38e-23
T = 295.15
Vi_th = np.arange(-1,1,0.001)

#plotting
def I(Vi_th):
    return Is1*(np.exp((q*Vi_th)/(k*T))-1)

plt.plot(Vi_th, I(Vi_th), 'g-', lw=2, label="Theoretical")
plt.title(r'Theoretical IV Curve of Diode with $I_{s}$ = 10e-12')
plt.xlabel('Voltage (V)')
plt.ylabel('Current (mA)')
plt.axhline(linewidth=1, color='k')
plt.axvline(linewidth=1, color='k')
plt.legend(bbox_to_anchor=(0.05, 0.95), loc=2, borderaxespad=0.)
plt.axis([-1,0.9,-1,8])
plt.grid(True)
plt.savefig('40.5_12A_IV_equation_plots_10e-12.png')
```
