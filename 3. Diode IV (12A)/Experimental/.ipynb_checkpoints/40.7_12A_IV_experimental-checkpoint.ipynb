```python
#importing packages
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
%matplotlib inline
```

```python
#loading data
dataFB = np.loadtxt("40.7_12A_IV_experimental_FB.txt")
dataRB = np.loadtxt("40.7_12A_IV_experimental_RB.txt")

#voltages in
Vi_FB = dataFB[:,0]
Vi_RB = dataRB[:,0]
#voltages accross
Vo_FB = dataFB[:,1]
Vo_RB = -dataRB[:,1]
#currents accross
Io_FB = dataFB[:,3]
Io_RB = -dataRB[:,2]*10**-3
```

```python
plt.plot(Vo_FB,Io_FB,'b-', lw=2, label="Experimental Forward Biased")
plt.plot(Vo_RB, Io_RB, 'r-', lw=2, label="Experimental Reverse Biased")
plt.title('Experimental IV Curve')
plt.xlabel('Voltage (V)')
plt.ylabel('Current (mA)')
plt.axhline(linewidth=1, color='k')
plt.axvline(linewidth=1, color='k')
plt.legend(bbox_to_anchor=(0.05, 0.95), loc=2, borderaxespad=0.)
plt.axis([-1,0.9,-1,8])
plt.grid(True)
plt.savefig('40.7_12A_IV_experimental_output.png')
```
